```python
url ='http://www.myanmarconstitutionaltribunal.org.mm/lawdatabase/en/advanced-search?title=&field_ll_law_no_value=&field_ll_publication_date_value[value][date]=&field_enforce_date_value[value][date]=&tid=All&body_value=&field_stable_unstable_value=All&field_choose_law_tid=All&tid_1=All&page=1'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
soup
```

```python
data = soup.find_all('div', attrs={'class':'post-text'})
```

```python
links = [x.find('a')['href'] for x in data]
```

```python
df = pd.DataFrame(links)
```
