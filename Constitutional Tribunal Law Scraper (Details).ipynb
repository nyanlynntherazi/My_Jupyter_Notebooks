```python
url = 'http://www.myanmarconstitutionaltribunal.org.mm/lawdatabase/my/law/996'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
title = soup.find('h1', attrs={'class':'title'}).text #Heading
```

```python
title = title.strip()
```

```python
label = [x.text for x in soup.find_all('div', attrs={'class':'field-label'})]
```

```python
label.extend(['file_name',':'])
```

```python
items = [y.text for y in soup.find_all('div', attrs={'class':'field-item even'})]
```
